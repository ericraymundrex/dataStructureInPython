# LinkedList

```python
class Node:
        def __init__(self,data=None,next=None):
                self.data=data
                self.next=next

class ll:
        def insert_value_from_list(self,the_list):
                self.head=None
                for data in the_list:
                        self.insert_at_the_end(data)

        def __init__(self):
                self.head=None
        
        def insert_in_the_beginning(self,data):
                node=Node(data,self.head)
                self.head=node

        def print_the_ll(self):
                if(self.head==None):
                        return print("The ll is empty")
                itr=self.head
                ll_str=''
                while(itr):
                        ll_str=ll_str+str(itr.data)+"-->"
                        itr=itr.next
                print(ll_str+"None")
        
        def insert_at_the_end(self,data):
                if self.head is None:
                        self.head=Node(data,None)
                        return
                itr=self.head
                while itr.next is not None:
                        itr=itr.next
                itr.next=Node(data,None)
        
        def get_the_length(self):
                count=0
                itr=self.head
                while itr:
                        count=count+1
                        itr=itr.next
                return count
        def insert_in_the_given_index(self,data,position):
                if(position==0):
                        self.insert_in_the_beginning(data)
                        return
                count=0
                itr=self.head
                while count>position and itr!=None:
                        itr=itr.next
                node=Node(data)
                node.next=itr.next
                itr.next=node
        def remove_a_node_from_the_list(self,index):
                if self.head is None:
                        print("There is no element in the ll")
                        return None
                if index<0 or index>self.get_the_length():
                        print("index outof bound")
                        return None
                
                count=0
                itr=self.head
                while itr is not None and count+1<index:
                        itr=itr.next
                        count=count+1
                itr.next=itr.next.next
        def search_for_the_node(self,data):
                itr=self.head
                count=0
                while itr:
                        if itr.data==data:
                                return count
                        count=count+1
                        itr=itr.next
                return "Element not found"
       
if __name__=="__main__":
        ll=ll()
        ll.insert_at_the_end(2)
        ll.insert_at_the_end(3)
        ll.insert_at_the_end(6)
        ll.insert_in_the_given_index(100,1)
        ll.remove_a_node_from_the_list(20)
        print(ll.search_for_the_node(9))
        ll.print_the_ll()
```

# doubly linkedlist

```python
class Node:
        def __init__(self,data=None,next=None,pre=None):
                self.data=data
                self.next=next
                self.pre=pre

class dll:
        def __init__(self):
                self.head=None
        def insert_in_the_begining(self,data):
                if self.head is None:
                        node=Node(data,self.head)
                        self.head=node
                node=Node(data,self.head)
                self.head.pre=node
                self.head=node

        def print_the_ddl(self):
                itr=self.head
                ddl_str=""
                while itr:
                        ddl_str=ddl_str+str(itr.data)+"<-->"
                        itr=itr.next
                print(ddl_str+"None")

        def insert_in_the_last(self,data):
                itr=self.head
                while itr.next is not None:
                        itr=itr.next
                node=Node(data,None,itr)
                itr.next=node
        def reverse_print(self):
                itr=self.head
                ddl_str=""
                while itr.next is not None:
                        itr=itr.next
                while itr:
                        ddl_str=ddl_str+str(itr.data)+"<-->"
                        itr=itr.pre
                print(ddl_str+"None")
        def count_the_number_of_elements(self):
                count=0
                itr=self.head
                while itr:
                        count=count+1
                        itr=itr.next
                return count
        
        def delete_the_node(self,index):
                if index<0 or index> self.count_the_number_of_elements():
                        print("The index outof bound")
                        return None
                if index is 0:
                        self.head=self.head.next
                        self.head.pre=None
                        return None
                
                itr=self.head
                position=0
                while itr is not None and position>index:
                        itr=itr.next
                        position=position+1
                itr.next=itr.next.next
                itr.next.pre=itr
        def pop(self):
                if self.head is None:
                        print("The dll is NULL")
                        return None
                itr=self.head
                while itr.next is not None:
                        itr=itr.next
                itr=itr.pre
                itr.next=None
if __name__=="__main__":
        dll=dll()
        dll.insert_in_the_begining(20)
        dll.insert_in_the_begining(30)
        dll.insert_in_the_begining(34)
        dll.insert_in_the_last(665)
        dll.insert_in_the_begining(33)
        dll.print_the_ddl()
        dll.reverse_print()
        dll.delete_the_node(2)
        dll.print_the_ddl()
        dll.pop()
        dll.print_the_ddl()
```
